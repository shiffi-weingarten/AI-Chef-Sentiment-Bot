```python
class ServiceManager:
    def __init__(self, api_key: str, system_prompt: str):
        self.api_key = api_key
        self.system_prompt = system_prompt
        self.conversation_history: List[Dict] = [] 
        self.client: Optional[genai.Client] = None 
        self.mock_mode = False # ברירת מחדל: לא במצב דמה

        # ניסיון חיבור ואימות (בדיקת שגיאות SSL / רשת)
        try:
            self.client = genai.Client(api_key=self.api_key)
            self.client.models.generate_content(
                model="gemini-2.5-flash", contents=[{"role": "user", "parts": [{"text": "בדיקה"}]}],
                config={"system_instruction": self.system_prompt}
            )
            print("✅ חברתי בהצלחה למודל Gemini.")
        except Exception as e:
            print(f"❌ שגיאת חיבור ל-API ({e}). הבוט פועל כעת במצב דמה (Mock Mode).")
            self.mock_mode = True
            
    #שומר תשובה ושאלה להסטוריה        
    def _save_message(self, role: str, text: str):
        self.conversation_history.append({'role': role, 'text': text})
        
    def send_message(self, user_question: str) -> str:
        self._save_message(role='user', text=user_question)

        #אם לא עובד 
        if self.mock_mode:
            response_index = (len(self.conversation_history) // 2) - 1 
            bot_answer = MOCK_RESPONSES[response_index % len(MOCK_RESPONSES)]
        #אם החיבור עבד בהצלחה    
        else:
            contents_for_api = [{"role": msg['role'], "parts": [{"text": msg['text']}]} for msg in self.conversation_history]
            try:
                response = self.client.models.generate_content(
                    model="gemini-2.5-flash", contents=contents_for_api,
                    config={"system_instruction": self.system_prompt}
                )
                bot_answer = response.text.strip()
            except Exception as e:
                bot_answer = f"אירעה שגיאת תקשורת במהלך שליחת השאלה: {e}"
                self.conversation_history.pop()
                
        self._save_message(role='model', text=bot_answer) 
        return bot_answer
```

```python
class ServiceManager(ServiceManager):
    
    def run_chat(self):
        print("🍳 ברוכים הבאים לשף הווירטואלי! (הבוט פועל במצב Mock אם יש שגיאת חיבור)")
        print('כדי לסיים את השיחה ולנתח את השירות, אנא כתוב "סיום".\n')
        
        while True:
            user_input = input("👤 משתמש: ")
            
            if user_input.lower() == 'סיום':
                print("\n✅ תודה רבה על השיחה! ניגש כעת לניתוח השירות.")
                self.finalize_service() 
                break
                
            bot_response = self.send_message(user_input)
            print(f"🤖🧑🏿‍🍳 שף וירטואלי: {bot_response}\n")


    def generate_ai_filename(self) -> str:
        if self.mock_mode:
            return "mock_chat_history.json"
        try:
            prompt = "הצע שם קובץ קצר בעברית (2-3 מילים) לשיחה הזו על אוכל. ענה רק בשם הקובץ בלי סיומת."
            response = self.client.models.generate_content(
                model="gemini-2.5-flash",
                contents=[{"role": "user", "parts": [{"text": prompt}]}]
            )
            clean_name = re.sub(r'[^\w\s]', '', response.text.strip()).replace(' ', '_')
            return f"{clean_name}.json"
        except:
            return "chef_chat_history.json"

    def save_history(self, filename: str):
        try:
            with open(filename, 'w', encoding='utf-8') as f:
                json.dump(self.conversation_history, f, ensure_ascii=False, indent=4)
            print(f"✅ היסטוריית השיחה נשמרה בהצלחה לקובץ: {filename}")
        except IOError as e:
            print(f"❌ שגיאת קובץ (שמירה): לא ניתן לשמור את הקובץ {filename}: {e}")
            raise
            
    def load_history(self, filename: str) -> List[Dict]:
        try:
            with open(filename, 'r', encoding='utf-8') as f:
                return json.load(f)
        except (FileNotFoundError, json.JSONDecodeError, IOError) as e:
            print(f"❌ שגיאת קובץ (טעינה): לא ניתן לטעון את הקובץ {filename}: {type(e).__name__}")
            return []

    def finalize_service(self):
        print("\n--- תהליך סיום והערכת שירות ---")
        filename = self.generate_ai_filename()
        
        if not self.conversation_history or (len(self.conversation_history) == 1 and self.conversation_history[0]['role'] == 'user'):
             print("אין היסטוריית שיחה לניתוח.")
             return
             
        try:
             self.save_history(filename) 
             self.analyze_sentiment_and_plot(filename) # עוברים לניתוח וגרפים
        except Exception:
             print("לא ניתן להמשיך לניתוח כי לא הצלחנו לשמור או לטעון.")
             return
```

```python
class ServiceManager(ServiceManager):
    
    def calculate_sentiment(self, text: str) -> float:
        # נורמליזציה: הסרת סימני פיסוק והפיכה לרישיות קטנות (דרישה)
        cleaned_text = re.sub(r'[^\w\s\-]', ' ', text.lower()) 
        words = [word for word in cleaned_text.split() if word]
        
        # חישוב סנטימנט באמצעות Dictionary Comprehension (דרישה)
        sentiment_scores = {
            word: SENTIMENT_DICT[word] 
            for word in words 
            if word in SENTIMENT_DICT
        }
        
        return sum(sentiment_scores.values())

    # --- דרישה: הצגת גרפים (Line & Bar Chart) ---
    def analyze_sentiment_and_plot(self, filename: str):
        
        loaded_history = self.load_history(filename)
        if not loaded_history:
            return

        df = pd.DataFrame(loaded_history)
        # חישוב ציון הסנטימנט לכל הודעה
        df['sentiment_score'] = df['text'].apply(self.calculate_sentiment)
        
        print("\n📊 יצירת גרפים:")

        # גרף 1: מגמת סנטימנט מצטבר (Line Chart)
        df['cumulative_sentiment'] = df['sentiment_score'].cumsum()
        plt.figure(figsize=(10, 5))
        plt.plot(df.index, df['cumulative_sentiment'], marker='o', color='blue')
        plt.title('גרף מגמת סנטימנט מצטבר של השיחה')
        plt.xlabel('מספר הודעה')
        plt.ylabel('ציון מצטבר')
        plt.axhline(0, color='gray', linestyle='-')
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.show()

        # גרף 2: השוואת ממוצע סנטימנט (Bar Chart)
        avg_sentiment = df.groupby('role')['sentiment_score'].mean()
        plt.figure(figsize=(7, 5))
        avg_sentiment.plot(kind='bar', color=['lightcoral', 'lightskyblue'])
        plt.title('השוואת ממוצע סנטימנט: משתמש מול שף')
        plt.ylabel('ממוצע ציון סנטימנט')
        plt.xticks(rotation=0)
        plt.grid(axis='y', linestyle='--', alpha=0.6)
        plt.show()
        
        # דוגמה לשימוש ב-Generator (דרישה מתקדמת)
        def negative_words_generator(history_df):
            for index, row in history_df.iterrows():
                if row['role'] == 'user' and row['sentiment_score'] < 0:
                     cleaned_text = re.sub(r'[^\w\s\-]', ' ', row['text'].lower())
                     for word in cleaned_text.split():
                         if word in SENTIMENT_DICT and SENTIMENT_DICT[word] < 0:
                             yield word
        
        negative_words = list(negative_words_generator(df))
        if negative_words:
            print(f"\n✅ דרישת Generator: נותחו {len(negative_words)} מילים שליליות.")
```
